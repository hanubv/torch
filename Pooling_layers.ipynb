```python
import torch
import torch.nn as nn
```

```python
#MaxPool-1D
#Syntax: torch.nn.MaxPool1D(kernal_size, stride=None, padding=0)

m = nn.MaxPool1d(3, stride = 2)
inputs = torch.randn(6,6)
print(inputs)
print("The tensor after applying Maxpool1D is \n")
outputs  = m(inputs)
print(outputs)
```

```python
m = nn.MaxPool2d(3, stride =1)
print(m)
inputs = torch.randn(4, 10, 4)
print(inputs)
outputs = m(inputs)
print(outputs)
```
