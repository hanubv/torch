```python
import torch
import torch.nn as nn
```

# **Conv1D**
Conv1d applies 1dimensional convolution on inputs layers composed of several input planes.
In the simplest case, the input of layer is (N, Cin, L) and ouput can be represented as (N, Cout, Lout) and representation can be showna as

out(N, Cout) = bias(Cout)+sum(weights(Cout, k)*input(Ni, k)

where:
N is input batch size

C is number of channels

*-operator is cross correlation or mulitplication operator and

L denoted length of the signal sequence


#Syntax:
torch.nn.Conv1d(int_channels, out_channels, kernal_size, stride=1, padding = 0, dilation = 0)

```python
m = nn.Conv1d(16, 33, 3, stride = 2)
print(m)
inputs = torch.randn(20, 16, 50)
#print(inputs)
print("\n")
ouput = m(inputs)
print(ouput)
```

```python
#CONV2D
#Syntax: torch.nn.Conv2d(in_channels, out_channels, kernal_size, stride)
#onv2d function with square size kernals
m = nn.Conv2d(16, 33, 3, stride = 2)
print(m)
inputs = torch.randn(20, 16, 50, 10)
print(inputs)
#output = m(inputs)
print(ouput)
```

```python
#Conv2d function with non square filters, unequal striding, and with padding
m = nn.Conv2d(16,33, (3,4), stride = (2,1), padding = (4,2))
print(m)
inputs = torch.randn(2,16, 6, 10)
#print(inputs)
outputs = m(inputs)
print(outputs)
```
